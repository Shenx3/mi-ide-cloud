# 🧹 Paso 3: Limpieza de datos
En este paso verificaremos y corregiremos los tipos de datos clave, y buscaremos valores inconsistentes en el dataset de transacciones de tarjetas de crédito.

- Convertir columnas de fecha y monto a su tipo correcto
- Verificar que los valores sean válidos (ejemplo: montos positivos, género válido, etc.)

```python
# Conversión de tipos de datos y chequeo de valores inconsistentes
import numpy as np

# Convertir columnas de fecha
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'], errors='coerce')

# Convertir monto a numérico
df['amt'] = pd.to_numeric(df['amt'], errors='coerce')

# Verificar montos negativos
print('Transacciones con monto negativo:', (df['amt'] < 0).sum())

# Verificar género válido
print('Valores únicos en género:', df['gender'].unique())

# Mostrar resumen de tipos de datos y primeras filas
print(df.dtypes)
df.head()
```

# 🔄 Paso 4: Corrección y transformación de tipos de datos
En este paso, nos aseguraremos de que las columnas clave tengan el tipo de dato correcto y crearemos nuevas variables útiles para el análisis de fraude.

- Confirmar que las fechas y montos están correctamente convertidos
- Crear variables adicionales si es necesario (por ejemplo, año de la transacción, edad del titular, etc.)

```python
# Confirmar tipos y crear variables adicionales
import pandas as pd
import numpy as np

# Confirmar tipos
print(df.dtypes)

# Crear columna: año de la transacción
df['trans_year'] = df['trans_date_trans_time'].dt.year

# Crear columna: edad del titular al momento de la transacción
df['edad'] = df['trans_date_trans_time'].dt.year - df['dob'].dt.year

# Mostrar primeras filas para verificar
df[['trans_date_trans_time', 'dob', 'trans_year', 'edad']].head()
```
